## Utils functions

```python
def describe_match(data_dictionary):
    '''
    
    Prints out the initial information about the parsed file, helpful for general orientation
    
    '''
    
    print(f'Initial file name: {data_dictionary["MatchId"]}')
    print(f'Client name: {data_dictionary["ClientName"]}')
    print(f'Map: {data_dictionary["MapName"]}')
    print(f'Total number of playback ticks: {data_dictionary["PlaybackTicks"]}')
    print(f'Tick Rate: {data_dictionary["TickRate"]}')
    print(f'Parse Rate: {data_dictionary["ParseRate"]}')
```
